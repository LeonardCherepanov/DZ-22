### 7. Проверить гипотезу, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом - 10 баллов

```python
# H0 сотрудники с высоким окладом проводят на работе одинаковое количество времени, что и сотрудники с низким окладом
# H1 сотрудники с высоким и низким окладами проводят разное время на работе

from scipy import stats as st
staff_low = data.loc[data['salary'] == 'low']
staff_high = data.loc[data['salary'] == 'high']

alpha = 0.05
result = st.ttest_ind(staff_high['average_montly_hours'], staff_low['average_montly_hours'], equal_var=False)
print(result)
if (result.pvalue < alpha):
    print('Отвергаем нулевую гипотезу')
else:
    print('Не отвергаем нулевую гипотезу')
```

**Вывод:** Гипотеза о том, что сотрудники с высоким окладом проводят на работе больше времени, чем сотрудники с низким окладом отклоняется.

### 9. Разделить данные на тестовую и обучающую выборки

#### Построить модель LDA, предсказывающую уволился ли сотрудник на основе имеющихся факторов (кроме department и salary)

#### Оценить качество модели на тестовой выборке - 20 баллов

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data[['satisfaction_level', 
                                                          'last_evaluation', 
                                                          'number_project', 
                                                          'average_montly_hours', 
                                                          'time_spend_company', 
                                                          'Work_accident', 
                                                          'promotion_last_5years']], data['left'], test_size=0.25)
```

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
lda.fit(X_train, y_train)
```

```python
lda.predict(X_test)
```

```python
# низкая точность модели
from sklearn.metrics import accuracy_score
accuracy_score(y_test, lda.predict(X_test))
```

```python
result = pd.DataFrame([y_test.values, lda.predict(X_test)]).T
result.head(50)
```
